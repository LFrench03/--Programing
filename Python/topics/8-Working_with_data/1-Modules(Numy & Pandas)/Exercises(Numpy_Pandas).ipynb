### Ejercicio 5
El fichero `cotizacion.csv` contiene las cotizaciones de las empresas del `IBEX35` con las siguientes columnas: 
```
Nombre (nombre de la empresa) 
Final (precio de la acción al cierre de bolsa) 
Máximo (precio máximo de la acción durante la jornada) 
Mínimo (precio mínimo de la acción durante la jornada)
Volumen (Volumen al cierre de bolsa)
Efectivo (capitalización al cierre en miles de euros)
```
Construir una función que construya un DataFrame a partir del un fichero con el formato anterior y devuelva otro DataFrame con el mínimo, el máximo y la media de dada columna.

````python
#Respuesta
def dataf(ruta):
    cotizacion = pd.read_csv(ruta)

    lista = cotizacion.columns  
    media = []
    min = []
    max = []
    for col in lista[1:]:
        media.append(np.mean(cotizacion[col]))
        min.append(np.min(cotizacion[col]))
        max.append(np.max(cotizacion[col]))
    data = {}
    data["Nombre"] = lista[1:]
    data["Media"] = media
    data["Min"] = min
    data["Max"] = max
    df = pd.DataFrame(data = data)
    df = df.set_index("Nombre")
    df.index.names = [None]
    return df
#Test
ruta = "data/cotizacion.csv"
df = dataf(ruta)
df.head()
````

### Ejercicio 14:
Write a  NumPy program to convert Centigrade degrees into Fahrenheit degrees. Centigrade values are stored in a NumPy array.
```

Values in Fahrenheit degrees:
[ 0. 12. 45.21 34. 99.91 32. ]

[-17.78 -11.11 7.34 1.11 37.73 0. ]
```

````python
#Respuesta:
F = np.array([0, 12, 45.21, 34, 99.91])
C = np.round((5 * F/9 - 5*32/9),2)
F = np.round(((9*C + (32*5))/5),2)
print("Grados Centigrados => {} \nGrados Fahrenheit => {}".format(F,C))

````

### Ejercicio 19:
Write a NumPy program to calculate the Euclidean distance:
$$d(P_{1},P_{n}) = \sqrt{\sum_{i=1}^{n}{{(x_{i}-x_{j})^{2}}}}$$
```
From Wikipedia:
In mathematics, the Euclidean distance or Euclidean metric is the "ordinary" straight-line distance between two points in Euclidean space. With this distance, Euclidean space becomes a metric space. The associated norm is called the Euclidean norm. Older literature refers to the metric as the Pythagorean metric.
Sample Output:
Euclidean distance: 5.196152422706632
```

````python
#Respuesta
'''Well, basically, this model it's just the square root of the sum of the distance of the points from eachother, squared. In Python terms, let's say you have something 
like:'''
def euclidean_distance(p1,p2):
    '''
    Args: Take two args values: {list or tuple that represents the coordinates of each p spot as p[0] = x & p[1] = 1}
    Out: Returns the result of the model to calculate the distance between this p1 & p2 spots
    '''
    #import numpy as np
    result = np.sqrt( (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2 )
    return f"d(P1,P2) = {result}"
#Test: 
plot1 = (1,3)
plot2 = (2,5)
print(euclidean_distance(plot1,plot2))

````
