### Ejercicio 1:
Escriba un programa Python para crear una clase que represente un c√≠rculo. Incluir m√©todos para calcular su √°rea y per√≠metro.

![](IMG/circulo.jpeg)

````python
from math import pi
class Circle(object): 
    def __init__(self, center = (0,0), radius = 1):
        if type(center) == tuple and (type(radius) == int or type(radius) == float):
            self.center = center
            self.radius = radius
        else:
            print("Center debe ser de tipo tupla y radius de tipo int o float")
    def __str__(self) -> str:
        return f"Centro = {self.center}\nRadio = {self.radius}"
    
    @property
    def diameter(self):
        return 2*self.radius
    
    @property
    def area(self):#Area
        return pi * (self.radius)**2
    
    @property
    def perimeter(self):#Perimetro
        return self.diameter * pi
    
#Prueba: 
c = Circle((3,3),5)
print(c)
print("Diametro =",c.diameter)
print("Area =",c.area)
print("Perimetro =",c.perimeter)
````

### Ejercicio 2:
Escriba un programa Python para crear una clase de persona. Incluya atributos como nombre, pa√≠s y fecha de nacimiento. Implementar un m√©todo para determinar la edad de la persona.

![](IMG/persona.jpeg)(referencia a [Persona](https://www.filmaffinity.com/es/film334326.html) JSJJASKJA juro que me ba√±o a veces)

````python
#Con datetime
from datetime import date
class Persona(object):
    def __init__(self, nombre, pais, fecha_de_nacimiento):
        if type(nombre) == str and type(pais) == str and type(fecha_de_nacimiento) == date:  
            self.nombre = nombre
            self.pais = pais
            self.fecha_de_nacimiento = fecha_de_nacimiento
        else:
            print("Los valores de entrada no corresponden con el tipo esperado.\ntype(nombre) == str and type(pais) == str and type(fecha_de_nacimiento) == date")
    def __str__(self) -> str:
        return "Nombre: {}\nPais: {}\nFecha de Nacimiento: {}".format(self.nombre, self.pais, self.fecha_de_nacimiento)

    @property
    def calcular_edad(self):
        today = date.today()
        years_old = today.year - self.fecha_de_nacimiento.year
        if (today.month, today.day) < (self.fecha_de_nacimiento.month, self.fecha_de_nacimiento.day):
            years_old -= 1
        return years_old

#Prueba: 
persona5 = Persona("Akira Kurusu/Ren Amamiya(Joker)", "Japon, Tokio", date(2000, 10, 2))
print(f"Nombre: {persona5.nombre}")
print(f"Pais: {persona5.pais}")
print(f"Fecha de Nacimiento: {persona5.fecha_de_nacimiento}")
edad = persona5.calcular_edad
print(f"Edad: {edad} a√±os")
````

### Ejercicio 3:
Escriba un programa Python para crear una clase de calculadora. **Incluir m√©todos para operaciones aritm√©ticas b√°sicas.**

![](IMG/calculadora.png)

````python
class Calculadora(object):
    def __init__(self):
        self.historial = []
        pass

    def suma(self, *numeros):
        resultado = numeros[0]
        for num in numeros[1:]:
            temp = resultado
            resultado += num
            self.historial.append(("+",temp,num,resultado))    
        return resultado

    def resta(self, *numeros):
        resultado = numeros[0]
        for num in numeros[1:]:
            temp = resultado
            resultado -= num
            self.historial.append(("-",temp,num,resultado))    
        return resultado

    def multiplicacion(self, *numeros):
        resultado = 1
        for num in numeros:
            temp = resultado
            resultado *= num
            self.historial.append(("*",temp,num,resultado))    
        return resultado

    def division(self, *numeros):
        divisor = numeros[1]
        resultado = numeros[0] / divisor
        self.historial.append(("/",numeros[0],divisor,resultado)) 
        for i in range(2, len(numeros)):
            temp = resultado
            divisor = numeros[i]
            resultado /= divisor
            self.historial.append(("/",temp,divisor,resultado)) 
        return resultado
    
    def __str__(self):
        return f"===Historial===\n{self.historial}"
    
    def registro(self,val):
        found = []
        for operation in self.historial:
            if val in operation:
                found.append(operation)
        if len(found)==0:
            return f"El valor {val} no se encontr√≥ en el historial"
        return f"===Registro===\nResultados: {found}"

# Ejemplo de uso
calculadora = Calculadora()

print("Suma:", calculadora.suma(5, 3, 2))
print("Resta:", calculadora.resta(10, 4, 3))
print("Multiplicaci√≥n:", calculadora.multiplicacion(6, 7, 8))
print("Divisi√≥n:", calculadora.division(20, 5, 2, 4))
print(calculadora)
print(calculadora.registro(5))
````

````python
class Array:
    def __init__(self, data):
        self.data = data

    def __add__(self, other):
        if isinstance(other, Array):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            # Manejo de arrays 1D y 2D
            if isinstance(self.data[0], (list, tuple, set)) and isinstance(other.data[0], (list,tuple,set)):  # Suponiendo que self.data y other.data son 2D
                result = [[self[i][j] + other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self.data[0], (int,float))) and (isinstance(other.data[0], (list,tuple,set))):
                result = [[self[j] + other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(self.data[0], (list,tuple,set))):
                result = [[self[i][j] + other[j] for j in range(len(other.data))] for i in range(len(self.data[0]))]
                return Array(result)
            else:  # Suponiendo que self.data y other.data son 1D
                result = [self[i] + other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x + other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operaci√≥n de suma no soportada con este tipo de dato")
        

    def __sub__(self, other):
        if isinstance(other, Array):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            # Manejo de arrays 1D y 2D
            if isinstance(self.data[0], list):  # Suponiendo que self.data y other.data son 2D
                result = [[self[i][j] - other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self.data[0], (int,float))) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i] - other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i][j] - other[j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            else:  # Suponiendo que self.data y other.data son 1D
                result = [self[i] - other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x - other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operaci√≥n de resta no soportada con este tipo de dato")

    def __mul__(self, other):
        if isinstance(other, Array):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            # Manejo de arrays 1D y 2D
            if isinstance(self.data[0], list):  # Suponiendo que self.data y other.data son 2D
                result = [[self[i][j] * other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self.data[0], (int,float))) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[j] * other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i][j] * other[j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            else:  # Suponiendo que self.data y other.data son 1D
                result = [self[i] * other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x * other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operaci√≥n de multiplicacion no soportada con este tipo de dato")
    
    def __matmul__(self, other):
        if isinstance(other,Array):
            if isinstance(self.data[0],(list,tuple,set)) and isinstance(other.data[0],(list,tuple,set)) and (len(other.data[0])>1):
                if (len(self.data)!= len(other.data[0])):
                    raise ValueError("Las matrices no pueden multiplicarse si el n√∫mero de filas de la primera no coincide con el n√∫mero de columnas de la segunda")
                result = []
                for i in range(len(self.data)):
                    result.append([])
                    for j in range(len(other.data[0])):
                        product = 0
                        for k in range(len(self.data[0])):
                            product+= self.data[i][k] * other.data[k][j]
                        result[i].append(product)
                return Array(result)                           
            elif isinstance(self.data[0],(int,float)) and isinstance(other.data[0],(list,tuple,set)): #Producto escalar por la izquierda
                if (len(self.data)!=len(other.data[0])):
                    raise ValueError("Las matrices no pueden multiplicarse si el n√∫mero de filas de la primera no coincide con el n√∫mero de columnas de la segunda")
                result = []
                n = len(self.data)
                m =  len(other.data)
                for i in range(n):
                    result.append([])
                    product = 0
                    for j in range(m):
                        product+= self.data[j] * other.data[j][i]
                    result[i].append(product)
                return Array(result)
            elif isinstance(self.data[0],(list,tuple,set)) and isinstance(other.data[0],(list,tuple,set)) and (len(other.data[0]) == 1): #Producto escalar por la derecha
                if (len(self.data)!=len(other.data)):
                    raise ValueError("Las matrices no pueden multiplicarse si el n√∫mero de fukas de la primera no coincide con el n√∫mero de columnas de la segunda")
                result = []
                n = len(self.data)
                m = len(other.data)
                for i in range(n):
                    result.append([])
                    product = 0    
                    for j in range(m):
                        product += self.data[j][i] * other.data[i][0]
                    result[i].append(product)
                return Array(result)
            else: 
                raise TypeError("Operaci√≥n de multiplicacion matricial no soportada con este tipo de dato")
        else:
            raise TypeError("Operaci√≥n de multiplicacion matricial no soportada con este tipo de dato")
        
    def __truediv__(self, other):
        if isinstance(other, (Array)):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            # Manejo de arrays 1D y 2D
            if isinstance(self[0], list):  # Suponiendo que self.data y other.data son 2D
                result = [[self[i][j] / other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self[0], (int,float))) and (isinstance(other[0], (list,tuple,Array))):
                result = [[self[i] / other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i][j] / other[j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            else:  # Suponiendo que self.data y other.data son 1D
                result = [self[i] / other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x / other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operaci√≥n de division no soportada con este tipo de dato")

    def __getitem__(self, index):
        if isinstance(index, int):
            return self.data[index]
        elif isinstance(index, slice):
            return Array(self.data[index])
        else:
            raise IndexError("√çndice fuera de rango")

    def __len__(self):
        return len(self.data)

    def __repr__(self):
        if isinstance(self.data[0], list):  # Caso de un array 2D
            out = '\n'.join(', '.join(map(str, row)) for row in self.data)
            out = str(out).replace(",", " ").replace("\n", "]\n[")
            return f"[[{out}]]"
        if isinstance(self.data, list):  # Caso de un array 1D
            out = ', '.join(map(str, self.data))
            out = str(out).replace(",", "")
            return f"[{out}]"

# Ejemplo:
one_dim_array = Array([1, 2, 3])
two_dim_array = Array([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
a = Array([[1, 2, 3], [4, 5, 6], [7, 8, 9]]) 
b = Array(([1],[2],[3]))
print("Array 1D:\n",one_dim_array,
    "\n===Suma de arrays 1D a si mismo===\n", one_dim_array + one_dim_array,
    "\n===Resta de array 1D a si mismo===\n", one_dim_array - one_dim_array,
    "\n===Producto simple de array 1D con el mismo===\n", one_dim_array * one_dim_array,
    "\n===Division de array 1D con el mismo===\n",one_dim_array / one_dim_array,
    "\n\nArray 2D:\n",two_dim_array,
    "\n===Suma de un array 2D con el mismo===\n", two_dim_array+ two_dim_array,
    "\n===Resta de un array 2D con el mismo===\n", two_dim_array - two_dim_array,
    "\n===Producto simple de un array 2D con el mismo===\n", two_dim_array * two_dim_array,
    "\n===Division de un array 2D con el mismo===\n", two_dim_array / two_dim_array,
    "\n===Producto matricial de un array 2D con el mismo===\n",two_dim_array @ two_dim_array,
    "\n\nArray 1D y Array 2D:\n",
    "\n===Suma===\n", one_dim_array + two_dim_array,
    "\n===Resta===\n", one_dim_array - two_dim_array,
    "\n===Producto simple===\n", one_dim_array * two_dim_array,
    "\n===Division===\n", one_dim_array / two_dim_array,
    "\n===Producto Matricial===\n", one_dim_array @ two_dim_array,
    "\n",a @ b)
    

````

````python
import math
class RationalNumber(object):
    '''
    Clase para trabajar con numeros racionales
    '''
    def __init__(self,n,d = 1):
        if type(n) is int and type(d) is int:
            self.numerator = n
            self.denominator = d
        else:
            print("El numerador y el denominador deben ser numeros enteros")
    
    def __str__(self):
        return (f"{self.numerator} / {self.denominator}")
    
    @property
    def mathFormat(self):
        from IPython.display import display, Latex
        display(Latex(f"${self.numerator} \\over {self.denominator}$"))

    @property
    def quotient(self):
        return self.numerator / self.denominator
    
    def isInfinite(self):
        return False if self.denominator != 0 else True
    
    def simplify(self):
        from numpy import gcd#Metodo de Numpy para hallar el mcd
        div = gcd(self.numerator, self.denominator)
        if div > 1:
            new_numerator = self.numerator / div
            new_denominator = self.denominator / div
            return ("Fraccion resultante simplificada: \n{} / {}".format(new_numerator,new_denominator))
        return "La fraccion es irreducible"
    
    @staticmethod
    def sum(p,q):
        new_num = (p.numerator * q.denominator) + (q.numerator * p.denominator)
        new_den = p.denominator * q.denominator
        print("{} / {} = {}".format(new_num,new_den,new_num / new_den))
    
    @staticmethod
    def substract(p,q):
        new_num = (p.numerator * q.denominator) - (q.numerator * p.denominator)
        new_den = p.denominator * q.denominator
        print("{} / {} = {}".format(new_num,new_den,new_num / new_den))
    
    @staticmethod
    def product(p,q):
        new_num = p.numerator * q.numerator
        new_den = p.denominator * q.denominator
        print("{} / {} = {}".format(new_num,new_den,new_num / new_den))
    
    @staticmethod
    def division(p,q):
        new_num = p.numerator * q.denominator
        new_den = p.denominator * q.numerator
        print("{} / {} = {}".format(new_num,new_den,new_num / new_den))

    @classmethod
    def random(cls):
        from numpy import random
        n = random.randint(1,100)
        d = random.randint(1,100)
        return cls(n,d)
    
    @classmethod
    def zero(cls):
        return cls(0)
    
    @classmethod
    def one(cls):
        return cls(1)
    
    @classmethod
    def fromRealNumber(cls,f):
        num = f
        den = 1
        d, i = math.modf(num)
        while d != 0:
            num *= 10
            den *= 10
            d, i = math.modf(num)
        num = int(num)
        den = int(den)
        return cls(num,den)
    
# Pruebas:
rational1 = RationalNumber(16, 4)
print(rational1)
print(rational1.isInfinite())  


# Simplificando la fracci√≥n
simplified = rational1.simplify()
print(simplified)  

rational2 = RationalNumber(5, 6)
rational2.mathFormat
rational3 = RationalNumber(7, 8)
rational3.mathFormat

# Suma
RationalNumber.sum(rational1, rational2)
# Resta
RationalNumber.substract(rational1, rational2)
# Producto
RationalNumber.product(rational1, rational2)
# Divisi√≥n
RationalNumber.division(rational1, rational2)

# Generando una fracci√≥n aleatoria
random_rational = RationalNumber.random()
random_rational.mathFormat

# Creando ceros y unos 
zero_rational = RationalNumber.zero()
one_rational = RationalNumber.one()
zero_rational.mathFormat  
one_rational.mathFormat

# Convertir un n√∫mero real a una fracci√≥n
fraction_from_real = RationalNumber.fromRealNumber(0.12) 
fraction_from_real.mathFormat
````

````python
class node:
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next

class linked_list:
    def __init__(self):
        self.head = None
    
    def add(self, data, place = "front"):
        '''
        Takes place parameter as the place where we're gonna introduce our new value on the linked list object we're handling.
            * 'front' as default for ordinary inplace at the front of the list
            * 'end' for adding to the end as a new final null node of the list               
        '''
        if place == 'front':
            self.head = node(data=data,next=self.head)
            return
        elif place == 'end': 
            if not self.head:
                self.head = node(data=data)
                return
            curr = self.head
            while curr.next:
                curr = curr.next
            curr.next = node(data=data)

    @property
    def is_empty(self):     #check if the linked list is empty
        return self.head == None
    
    def delete_node(self, key=None): 
        '''
        Takes key as the parameter that defines if we're gonna delete the head element as default by a None value
        or we're gonna specify some element in particular located anywhere in the linked list by checking every reference of every node since the head to the tail.
        '''
        curr = self.head
        prev = None
        if key is not None:
            while curr and curr.data != key:
                prev = curr
                curr = curr.next
        else:
            while curr and curr.next is not None:
                prev = curr
                curr = curr.next
        if prev is None:
            self.head = curr.next
        elif curr:
            prev.next = curr.next
            curr.next = None

    @property
    def get_last_node(self):    #get the last node of the linked list
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        return temp.data
    
    @property #print as linked list
    def print(self):
        node = self.head
        while(node != None):
            print(node.data, end=" => ")
            node = node.next
    
    def __len__(self):  #return de amount of elements in the linked list
        count = 0
        if self.head:
            count +=1
            node = self.head
            while (node.next is not None):
                node = node.next
                count+=1
        return count
    
    def find(self, data):   #search element 'data'
        node = self.head
        while (node != None):
            if node.data == data:
                return node.data == data
            node = node.next
        return node == data
    
    def __str__(self):
        if self.is_empty:
            return f"{None}"
        node = self.head
        string = "["
        while (node != None):
            string += f"{node.data}, "
            node = node.next
        string = string[:-2] + "]"
        return f"{string}"

s = linked_list() 
print(s.is_empty) 
s.add(5) 
s.add(8, place='end') 
s.add(9) 
s.delete_node(8)
print(len(s))
print(s.find(5))
print(s)
s.print

````

````python
class Stack(linked_list):
    def __init__(self):
        linked_list.__init__(self)

    def push(self, data):
        """
        Add a new element at the end of the stack.
        """
        self.add(data, place='end')

    @property
    def pop(self):
        """
        Delete and returns the tail element of the stack. If the stack is empty returns `None`
        """
        if self.is_empty:
            return None
        else:
            data = self.get_last_node
            self.delete_node(data)
            return data
    
#Test:    
stack = Stack()
print(stack.is_empty) 
stack.push(5) 
stack.push(8) 
stack.push(9) 
print(len(stack))
print(stack.find(5))
print(stack)
print(stack.print)
print(stack.pop)
print(stack.print)
````

Una cola con prioridad es una estructura de datos similar a una cola regular, pero cada elemento tiene una prioridad asociada, y los elementos se desenfilan en orden de prioridad, no en orden de llegada. El metodo add de la cola ademas de recibir el valor recibir√° la prioridad de este valor y al hacer dequee se devolveran los nodos que mas valor tengan primero y luego se ordenar√° por orden de llegada a la cola

````python
class Node:
    def __init__(self, data=None, priority=None, next=None):
        self.data = data
        self.priority = priority
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
    
    
    @property
    def is_empty(self):     #check if the linked list is empty
        return self.head == None
    
    def add_ordered(self, data, priority):
        new_node = Node(data=data, priority=priority)
        if self.head is None or priority >= self.head.priority:
            new_node.next = self.head
            self.head = new_node
        else:
            current = self.head
            while current.next and current.next.priority < priority:
                current = current.next
            new_node.next = current.next
            current.next = new_node
            # Ordena 
            current = self.head
            while current.next:
                if current.priority < current.next.priority:
                    temp = current.next
                    current.next = current.next.next
                    temp.next = self.head
                    self.head = temp
                else:
                    current = current.next

    def delete_node(self, key=None):
        curr = self.head
        prev = None
        if key is not None:
            while curr and curr.data != key:
                prev = curr
                curr = curr.next
        else:
            while curr and curr.next is not None:
                prev = curr
                curr = curr.next
        if prev is None:
            self.head = curr.next
        elif curr:
            prev.next = curr.next
            curr.next = None

    def __len__(self):
        count = 0
        if self.head:
            count += 1
            node = self.head
            while node.next is not None:
                node = node.next
                count += 1
        return count

    def find(self, data):
        node = self.head
        while node:
            if node.data == data:
                return True
            node = node.next
        return False

    def __str__(self):
        if self.is_empty:
            return "[]"
        node = self.head
        string = "[" + str(node.data)
        while node.next:
            node = node.next
            string += ", " + str(node.data)
        string += "]"
        return string

class PriorityQueue(LinkedList):
    def add(self, data, priority):
        self.add_ordered(data, priority)

    def dequeue(self):
        if self.is_empty:
            raise Exception("The Queue is empty")
        data = self.head.data
        self.delete_node(data)
        return data

# Uso de PriorityQueue
pq = PriorityQueue()
pq.add("Task 1", 2)
pq.add("Task 2", 3)
pq.add("Task 3", 1)
pq.add("Task 4", 1)
print(pq)  # Deber√≠a imprimir la cola ordenada por prioridad
print(pq.dequeue())  # Deber√≠a devolver "Task 2" ya que tiene la mayor prioridad
print(pq)  # Deber√≠a mostrar la cola despu√©s de eliminar "Task 2"
````

````python
class Arbol:
    def __init__(self, elemento):
        self.hijos = []
        self.elemento = elemento

    def agregar_elemento(self, elemento, elemento_padre):
        sub_arbol = self.buscar_sub_arbol(elemento_padre)
        if sub_arbol is not None:
            sub_arbol.hijos.append(elemento)
        else:
            print(f"No se encontr√≥ el elemento {elemento_padre} en el √°rbol.")

    def buscar_sub_arbol(self, elemento):   #DFS
        if self.elemento == elemento:
            return self
        for hijo in self.hijos:
            encontrado = hijo.buscar_sub_arbol(elemento)
            if encontrado is not None:
                return encontrado
        return None

    def profundidad(self):
        if len(self.hijos) == 0:
            return 1
        else:
            return 1 + max((hijo.profundidad() for hijo in self.hijos), default=0)

    def calcular_altura(self):
        """Calcula la altura del √°rbol."""
        return self.profundidad()

    def contiene_valor(self, valor):
        """Verifica si un valor est√° presente en el √°rbol."""
        if self.elemento == valor:
            return True
        for hijo in self.hijos:
            if hijo.contiene_valor(valor):
                return True
        return False

#Test: 
abuela = Arbol("Jacqueline Gurney")
patty = Arbol("Patty Bouvier")
selma = Arbol("Selma Bouvier")
ling = Arbol("Ling Bouvier")
marge = Arbol("Marge Bouvier")
bart = Arbol("Bart Simpson")
lisa = Arbol("Lisa Simpson")
maggie = Arbol("Maggie Simpson")

abuela.agregar_elemento(patty, "Jacqueline Gurney")
abuela.agregar_elemento(selma, "Jacqueline Gurney")
abuela.agregar_elemento(marge, "Jacqueline Gurney")

selma.agregar_elemento(ling, "Selma Bouvier")

marge.agregar_elemento(bart, "Marge Bouvier")
marge.agregar_elemento(lisa, "Marge Bouvier")
marge.agregar_elemento(maggie, "Marge Bouvier")

print("Altura del √°rbol:", abuela.calcular_altura())
print("Contiene el valor 'Bart Simpson'?", abuela.contiene_valor('Bart Simpson')) 
````

Para la anterior representacion de algunas ciudades de Argentina, asuma que el peso de las aristas significa la cantidad de combustible que necesita para ir de un lugar a otro, determine si hay alguna forma de llegar a Medoza desde San Pedro gastando un maximo de 1000 litros de combustible

````python
#Ejemplo grafo:
class Grafo:
    def __init__(self):
        self.grafo = {}

    def __str__(self):
        return str(self.grafo)

    def agregar_arista(self, origen, destino, peso):
        if origen not in self.grafo:
            self.grafo[origen] = []
        if destino not in self.grafo:
            self.grafo[destino] = []
        self.grafo[origen].append((destino, peso))
        self.grafo[destino].append((origen, peso))

    def buscar_ruta(self, inicio, fin, max_combustible):
        def dfs(ciudad_actual, destino, combustible_restante, ruta):
            if combustible_restante < 0:
                return None
            if ciudad_actual == destino:
                return ruta
            for vecino, peso in self.grafo.get(ciudad_actual, []):
                if vecino not in ruta:  # Evitar ciclos
                    resultado = dfs(vecino, destino, combustible_restante - peso, ruta + [vecino])
                    if resultado is not None:
                        return resultado
            return None

        return dfs(inicio, fin, max_combustible, [inicio])


grafo = Grafo()
grafo.agregar_arista("C√≥rdoba", "Villa Mar√≠a", 150)
grafo.agregar_arista("Villa Mar√≠a", "San Luis", 350)
grafo.agregar_arista("Villa Mar√≠a", "Rosario", 245)
grafo.agregar_arista("Rosario", "San Pedro", 160)
grafo.agregar_arista("San Pedro", "Buenos Aires", 175)
grafo.agregar_arista("Buenos Aires", "Bah√≠a Blanca", 660)
grafo.agregar_arista("Bah√≠a Blanca", "San Luis", 800)
grafo.agregar_arista("San Luis", "Mendoza", 260)
print(grafo)
# Buscar la ruta
ruta = grafo.buscar_ruta("San Pedro", "Mendoza", 1050)
print("Ruta encontrada:", ruta)
````
