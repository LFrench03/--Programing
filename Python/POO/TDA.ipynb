# <div style="text-align:center"> <l style= "color: orange">*Linked List (Lista Enlazada)*</l></div>
<div style="text-align:center"> <img src="IMG/linkedlist.png" scale=5.0/></div>

---

Métodos:

- `add`
- `delete`
- `isempty`
- `get_end`
- `print`
- `find`
- `__len__`

- `__str__`


````python
class node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class LinkedList:
    def __init__(self):
        self.head = None
    
    def add(self, data, place = "front"):
        if place == "front":
            self.head = node(data=data, next=self.head)
            return
        elif place == "end":
            if not self.head:
                self.head = node(data=data)
                return
            curr = self.head
            while curr.next:
                curr = curr.next
            curr.next = node(data=data)

    def delete(self, key = None):
        prev = None
        curr = self.head
        if key is not None:
            while curr and curr.data != key:
                prev = curr
                curr = curr.next
        elif key is None:
            while curr and curr.next !=None:
                prev = curr
                curr = curr.next
        if prev is None:
            self.head = curr.next
        else:
            prev.next = curr.next
            curr.next = None

    @property
    def isempty(self):
        return self.head == None
    
    @property
    def get_end(self):
        temp = self.head
        while temp.next:
            temp = temp.next
        return temp.data
    
    @property
    def print(self):
        node = self.head
        while(node!=None):
            print(node.data, end="=>")
            node = node.next
    
    def __len__(self):
        count = 0
        if self.head:
            count +=1
            node = self.head
            while node.next:
                node = node.next
                count+=1
        return count
    
    def find(self, value):
        node = self.head
        while (node.next != None):
            if node.data == value:
                return node.data == value
            node = node.next
        return node == value
    
    def __str__(self):
        if self.is_empty:
            return f"{None}"
        node = self.head
        string = "["
        while (node != None):
            string += f"{node.data}, "
            node = node.next
        string = string[:-2] + "]"
        return f"{string}"
    
    
````

# <div style="text-align:center"> <l style= "color: orange">*Queue (Cola)*</l></div>
<div style="text-align:center"> <img src="IMG/cola.png" scale=7.0/></div>

---

Hereda de `LinkedList` y tiene como métodos:

- `enqueue`
- `dequeue`
- `rear`
- `front`



````python
class Queue(LinkedList):
    def __init__(self):
        LinkedList.__init__(self)
    
    def enqueue(self, data): 
        self.add(data, "end")
    
    def dequeue(self):
        if self.isempty:
            return None
        else:
            data = self.head.data
            self.delete(data)
            return data
    
    @property
    def front(self):
        if self.isempty == False:
            return None
        return self.head.data
            
    
    @property
    def rear(self):
        if self.isempty:
            return None
        data = self.get_end
        return data
    
````

# <div style="text-align:center"> <l style= "color: orange">*Stack (Pila)*</l></div>
<div style="text-align:center"> <img src="IMG/stack.png" width = 750px height= 400px/></div>

---

Hereda de `LinkedList` y tiene como métodos:
\
- `push`
- `pop`
- `top`


````python
class Stack(LinkedList):
    def __init__(self):
        LinkedList.__init__(self)
    
    def push(self,data):
        self.add(data, "end")

    def pop(self):
        if self.isempty:
            return None
        data = self.get_end
        self.delete(data)
        return data
    
    def top(self):
        if self.isempty:
            return None
        data = self.get_end
        return data
        
````

# <div style="text-align:center"> <l style= "color: orange">*PriorityQueue (Cola con prioridad)*</l></div>
<div style="text-align:center"> <img src="IMG/pqueue.jpg" width = 850px height= 400px/></div>

---

Una `cola con prioridad` es una estructura de datos similar a una `cola` regular, pero cada elemento tiene una `prioridad` asociada, y los elementos se desenfilan en orden de prioridad, no en orden de llegada. 

- El metodo `add` de la cola ademas de recibir el valor recibirá la prioridad de este valor y al hacer `dequee` se devolveran los nodos que mas valor tengan primero y luego se ordenará por orden de llegada a la cola

```
pq = PriorityQueue()
pq.add("Task 1", 2)
pq.add("Task 2", 3)
pq.add("Task 3", 1)
pq.add("Task 4", 1)

print(pq.dequeue())  # Output: Task 2
print(pq.dequeue())  # Output: Task 1
print(pq.dequeue())  # Output: Task 3
print(pq.dequeue())  # Output: Task 4


```



````python
class node:
    def __init__(self, data, priority, next=None):
        self.data = data
        self.next = next
        self.priority = priority

class PriorityQueue:
    def __init__(self):
        self.head = None
    
    def add(self, data, priority):
        new_node = node(data=data, priority=priority)
        if self.head == None:
            self.head = new_node

        elif new_node.priority > self.head.priority:
            new_node.next = self.head
            self.head = new_node
            
        elif new_node.priority <= self.head.priority:
            current = self.head
            while current.next and new_node.priority <= current.next.priority:
                current = current.next
            new_node.next = current.next
            current.next = new_node

    @property
    def dequeue(self):
        if self.head is None:
            return None
        data = self.head.data
        self.head = self.head.next
        return data
    
    def __str__(self):
        if self.head == None:
            return "[]"
        node = self.head
        string = "[" + str(node.data)
        while node.next:
            node = node.next
            string += ", " + str(node.data)
        string += "]"
        return string

        
    
pq = PriorityQueue()
pq.add("Task 1", 2)
pq.add("Task 2", 3)
pq.add("Task 3", 1)
pq.add("Task 4", 1)
print(pq)  
print(pq.dequeue)  
print(pq)  
````
