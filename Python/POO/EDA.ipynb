````python
class Array:
    def __init__(self, data):
        self.data = data

    def __add__(self, other):
        if isinstance(other, Array):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            if isinstance(self.data[0], (list, tuple, set)) and isinstance(other.data[0], (list,tuple,set)):
                result = [[self[i][j] + other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self.data[0], (int,float))) and (isinstance(other.data[0], (list,tuple,set))):
                result = [[self[j] + other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(self.data[0], (list,tuple,set))):
                result = [[self[i][j] + other[j] for j in range(len(other.data))] for i in range(len(self.data[0]))]
                return Array(result)
            else:  # Suponiendo que self.data y other.data son 1D
                result = [self[i] + other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x + other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operación de suma no soportada con este tipo de dato")
        

    def __sub__(self, other):
        if isinstance(other, Array):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            if isinstance(self.data[0], list):  
                result = [[self[i][j] - other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self.data[0], (int,float))) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i] - other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i][j] - other[j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            else:  # Suponiendo que self.data y other.data son 1D
                result = [self[i] - other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x - other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operación de resta no soportada con este tipo de dato")

    def __mul__(self, other):
        if isinstance(other, Array):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            if isinstance(self.data[0], list): 
                result = [[self[i][j] * other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self.data[0], (int,float))) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[j] * other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i][j] * other[j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            else:
                result = [self[i] * other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x * other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operación de multiplicacion no soportada con este tipo de dato")
    
    def __matmul__(self, other):
        if isinstance(other,Array):
            if isinstance(self.data[0],(list,tuple,set)) and isinstance(other.data[0],(list,tuple,set)) and (len(other.data[0])>1):
                if (len(self.data)!= len(other.data[0])):
                    raise ValueError("Las matrices no pueden multiplicarse si el número de filas de la primera no coincide con el número de columnas de la segunda")
                result = []
                for i in range(len(self.data)):
                    result.append([])
                    for j in range(len(other.data[0])):
                        product = 0
                        for k in range(len(self.data[0])):
                            product+= self.data[i][k] * other.data[k][j]
                        result[i].append(product)
                return Array(result)                           
            elif isinstance(self.data[0],(int,float)) and isinstance(other.data[0],(list,tuple,set)): #Producto escalar por la izquierda
                if (len(self.data)!=len(other.data[0])):
                    raise ValueError("Las matrices no pueden multiplicarse si el número de filas de la primera no coincide con el número de columnas de la segunda")
                result = []
                n = len(self.data)
                m =  len(other.data)
                for i in range(n):
                    result.append([])
                    product = 0
                    for j in range(m):
                        product+= self.data[j] * other.data[j][i]
                    result[i].append(product)
                return Array(result)
            elif isinstance(self.data[0],(list,tuple,set)) and isinstance(other.data[0],(list,tuple,set)) and (len(other.data[0]) == 1): #Producto escalar por la derecha
                if (len(self.data)!=len(other.data)):
                    raise ValueError("Las matrices no pueden multiplicarse si el número de fukas de la primera no coincide con el número de columnas de la segunda")
                result = []
                n = len(self.data)
                m = len(other.data)
                for i in range(n):
                    result.append([])
                    product = 0    
                    for j in range(m):
                        product += self.data[j][i] * other.data[i][0]
                    result[i].append(product)
                return Array(result)
            else: 
                raise TypeError("Operación de multiplicacion matricial no soportada con este tipo de dato")
        else:
            raise TypeError("Operación de multiplicacion matricial no soportada con este tipo de dato")
        
    def __truediv__(self, other):
        if isinstance(other, (Array)):
            if len(self.data)!= len(other.data):
                raise ValueError("Los arrays deben tener la misma longitud o uno de ellos un array de 1D con el numero de elementos iguales al numero de filas de otro array")
            if isinstance(self[0], list):  
                result = [[self[i][j] / other[i][j] for j in range(len(self.data[0]))] for i in range(len(self.data))]
                return Array(result)
            if (isinstance(self[0], (int,float))) and (isinstance(other[0], (list,tuple,Array))):
                result = [[self[i] / other[i][j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            if isinstance(other.data[0], (int,float)) and (isinstance(other.data[0], (list,tuple,Array))):
                result = [[self[i][j] / other[j] for j in range(len(self.data))] for i in range(len(other.data[0]))]
                return Array(result)
            else:  
                result = [self[i] / other[i] for i in range(len(self.data))]
                return Array(result)
        elif isinstance(other, (int, float)):
            result = [x / other for x in self.data]
            return Array(result)
        else:
            raise TypeError("Operación de division no soportada con este tipo de dato")

    def __getitem__(self, index):
        if isinstance(index, int):
            return self.data[index]
        elif isinstance(index, slice):
            return Array(self.data[index])
        else:
            raise IndexError("Índice fuera de rango")

    def __len__(self):
        return len(self.data)

    def __repr__(self):
        if isinstance(self.data[0], (list,tuple,set)):  
            out = '\n'.join(', '.join(map(str, row)) for row in self.data)
            out = str(out).replace(",", " ").replace("\n", "]\n[")
            return f"[[{out}]]"
        if isinstance(self.data, (list,tuple,set)):
            out = ', '.join(map(str, self.data))
            out = str(out).replace(",", "")
            return f"[{out}]"

'''
#Test
one_dim_array = Array([1, 2, 3])
two_dim_array = Array([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
vector_col = Array(([1],[2],[3]))
print("Array 1D:\n",one_dim_array,
    "\n===Suma de arrays 1D a si mismo===\n", one_dim_array + one_dim_array,
    "\n===Resta de array 1D a si mismo===\n", one_dim_array - one_dim_array,
    "\n===Producto simple de array 1D con el mismo===\n", one_dim_array * one_dim_array,
    "\n===Division de array 1D con el mismo===\n",one_dim_array / one_dim_array,
    "\n\nArray 2D:\n",two_dim_array,
    "\n===Suma de un array 2D con el mismo===\n", two_dim_array+ two_dim_array,
    "\n===Resta de un array 2D con el mismo===\n", two_dim_array - two_dim_array,
    "\n===Producto simple de un array 2D con el mismo===\n", two_dim_array * two_dim_array,
    "\n===Division de un array 2D con el mismo===\n", two_dim_array / two_dim_array,
    "\n===Producto matricial de un array 2D con el mismo===\n",two_dim_array @ two_dim_array,

    "\n\nArray 1D y Array 2D:\n",
    "\n===Suma===\n", one_dim_array + two_dim_array,
    "\n===Resta===\n", one_dim_array - two_dim_array,
    "\n===Producto simple===\n", one_dim_array * two_dim_array,
    "\n===Division===\n", one_dim_array / two_dim_array,
    "\n===Producto Escalar por la izquierda===\n", one_dim_array @ two_dim_array,
    "\n===Producto Escalar por la derecha===\n",two_dim_array @ vector_col)
'''
````

# <div style="text-align:center"> <l style= "color: orange">*Linked List (Lista Enlazada)*</l></div>
<div style="text-align:center"> <img src="IMG/linkedlist.png" scale=5.0/></div>

---

Métodos:

- `add`
- `delete`
- `isempty`
- `get_end`
- `print`
- `find`
- `__len__`

- `__str__`


````python
class node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

class LinkedList:
    def __init__(self):
        self.head = None
    
    def add(self, data, place = "front"):
        if place == "front":
            self.head = node(data=data, next=self.head)
            return
        elif place == "end":
            if not self.head:
                self.head = node(data=data)
                return
            curr = self.head
            while curr.next:
                curr = curr.next
            curr.next = node(data=data)

    def delete(self, key = None):
        prev = None
        curr = self.head
        if key is not None:
            while curr and curr.data != key:
                prev = curr
                curr = curr.next
        elif key is None:
            while curr and curr.next !=None:
                prev = curr
                curr = curr.next
        if prev is None:
            self.head = curr.next
        else:
            prev.next = curr.next
            curr.next = None

    @property
    def isempty(self):
        return self.head == None
    
    @property
    def get_end(self):
        temp = self.head
        while temp.next:
            temp = temp.next
        return temp.data
    
    @property
    def print(self):
        node = self.head
        while(node!=None):
            print(node.data, end="=>")
            node = node.next
    
    def __len__(self):
        count = 0
        if self.head:
            count +=1
            node = self.head
            while node.next:
                node = node.next
                count+=1
        return count
    
    def find(self, value):
        node = self.head
        while (node.next != None):
            if node.data == value:
                return node.data == value
            node = node.next
        return node == value
    
    def __str__(self):
        if self.is_empty:
            return f"{None}"
        node = self.head
        string = "["
        while (node != None):
            string += f"{node.data}, "
            node = node.next
        string = string[:-2] + "]"
        return f"{string}"
    
    
````

# <div style="text-align:center"> <l style= "color: orange">*Queue (Cola)*</l></div>
<div style="text-align:center"> <img src="IMG/cola.png" scale=7.0/></div>

---

Hereda de `LinkedList` y tiene como métodos:

- `enqueue`
- `dequeue`
- `rear`
- `front`



````python
class Queue(LinkedList):
    def __init__(self):
        LinkedList.__init__(self)
    
    def enqueue(self, data): 
        self.add(data, "end")
    
    def dequeue(self):
        if self.isempty:
            return None
        else:
            data = self.head.data
            self.delete(data)
            return data
    
    @property
    def front(self):
        if self.isempty == False:
            return None
        return self.head.data
            
    
    @property
    def rear(self):
        if self.isempty:
            return None
        data = self.get_end
        return data
    
````

# <div style="text-align:center"> <l style= "color: orange">*Stack (Pila)*</l></div>
<div style="text-align:center"> <img src="IMG/stack.png" width = 750px height= 400px/></div>

---

Hereda de `LinkedList` y tiene como métodos:

- `push`
- `pop`
- `top`


````python
class Stack(LinkedList):
    def __init__(self):
        LinkedList.__init__(self)
    
    def push(self,data):
        self.add(data, "end")

    def pop(self):
        if self.isempty:
            return None
        data = self.get_end
        self.delete(data)
        return data
    
    def top(self):
        if self.isempty:
            return None
        data = self.get_end
        return data
        
````

# <div style="text-align:center"> <l style= "color: orange">*PriorityQueue (Cola con prioridad)*</l></div>
<div style="text-align:center"> <img src="IMG/pqueue.jpg" width = 850px height= 400px/></div>

---

Una `cola con prioridad` es una estructura de datos similar a una `cola` regular, pero cada elemento tiene una `prioridad` asociada, y los elementos se desenfilan en orden de prioridad, no en orden de llegada. 

- El metodo `enqueue` de la cola ademas de recibir el valor recibirá la prioridad de este valor y al hacer `dequee` se devolveran los nodos que mas valor tengan primero y luego se ordenará por orden de llegada a la cola

```python
pq = PriorityQueue()
pq.enqueue("Task 1", 2)
pq.enqueue("Task 2", 3)
pq.enqueue("Task 3", 1)
pq.enqueue("Task 4", 1)

print(pq.dequeue())  # Output: Task 2
print(pq.dequeue())  # Output: Task 1
print(pq.dequeue())  # Output: Task 3
print(pq.dequeue())  # Output: Task 4
```



````python
class node:
    def __init__(self, data, priority, next=None):
        self.data = data
        self.next = next
        self.priority = priority

class PriorityQueue:
    def __init__(self):
        self.head = None
    
    def enqueue(self, data, priority):
        new_node = node(data=data, priority=priority)
        if self.head == None:
            self.head = new_node

        elif new_node.priority > self.head.priority:
            new_node.next = self.head
            self.head = new_node
            
        elif new_node.priority <= self.head.priority:
            current = self.head
            while current.next and new_node.priority <= current.next.priority:
                current = current.next
            new_node.next = current.next
            current.next = new_node

    @property
    def dequeue(self):
        if self.head is None:
            return None
        data = self.head.data
        self.head = self.head.next
        return data
    
    def __str__(self):
        if self.head == None:
            return "[]"
        node = self.head
        string = "[" + str(node.data)
        while node.next:
            node = node.next
            string += ", " + str(node.data)
        string += "]"
        return string

        
    
pq = PriorityQueue()
pq.enqueue("Task 1", 7)
pq.enqueue("Task 2", 3)
pq.enqueue("Task 3", 1)
pq.enqueue("Task 4", 5)
print(pq)  
print(pq.dequeue)  
print(pq)   
````

# <div style="text-align:center"> <l style= "color: orange">*Trees (Árboles)*</l></div>
<div style="text-align:center"> <img src="IMG/tree.png" scale=5.0/></div>

---
Componentes:

- `Raiz` (si contiene hijos)
- `Hojas` (raices que no contienen hijos)
- `Hijos` (cada hijo se puede considerar como la raiz u hoja de un sub árbol)
-  `Profundidad`:  Es la altura determinada por cada salto generacional desde el nodo raiz

Desde un enfoque `recursivo`, con los métodos:

- `agregar_elemento`
- `buscar_sub_arbol` & `contiene_valor`
- `profundidad` & `calcular_altura`

````python
class Tree:
    def __init__(self, raiz):
        self.raiz = raiz
        self.hijos = []

    def agregar_elemento(self, elemento, padre):
        sub_arbol = self.buscar_sub_arbol(padre)
        if sub_arbol is not None:
            sub_arbol.hijos.append(elemento)
        else:
            print("El elemento padre {} no se encuentra en el árbol".format(padre))

    def buscar_sub_arbol(self, elemento):
        if self.raiz ==  elemento:
            return self
        for hijo in self.hijos:
            encontrado = hijo.buscar_sub_arbol(elemento)
            if encontrado is not None:
                return True
        return None
    
    def contiene_valor(self, valor):
        if self.raiz == valor:
            return True
        for hijo in self.hijos:
            contiene = hijo.contiene_valor(valor)
            if contiene:
                return True
        return False
    
    @property
    def profundidad(self):
        if len(self.hijos) == 0:
            profundidad = 1
        if len(self.hijos) > 0:
            return 1 + max((self.profundidad for hijo in self.hijos), default=0)

    @property    
    def calcular_altura(self):
        return self.profundidad
````

# <div style="text-align:center"> <l style= "color: orange">*Binary Tree (Árbol Binario)*</l></div>
<div style="text-align:center"> <img src="IMG/binary tree search.png" scale=5.0/></div>

---
Métodos:

- `Insertar`
- `Buscar`

````python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.raiz = None

    def Insertar(self, elemento):
        if self.raiz is None:
            self.raiz = Node(elemento)
        else:
            self.Insertar_(elemento, self.raiz)
    def Insertar_(self,elemento,node_actual):
        if elemento < node_actual.value:
            if node_actual.left is None:
                node_actual.left = Node(elemento)
            else:
                self.Insertar_(elemento, node_actual.left)
        elif elemento > node_actual.value:
            if node_actual.right is None:
                node_actual.right = Node(elemento)
            else:
                self.Insertar_(elemento, node_actual.right)  
        else:
            raise Exception("No se admiten valores duplicados en el arbol binario")
    
    def Buscar(self, elemento):
            return self.Buscar_(elemento, self.raiz)
    
    def Buscar_(self, elemento, node_actual):
        if node_actual is None:
            return False
        elif elemento < node_actual.value:
            return self.Buscar_(elemento, node_actual.left)
        elif elemento > node_actual.value:
            return self.Buscar_(elemento, node_actual.right)
        else:
            return True
````

# <div style="text-align:center"> <l style= "color: orange">*First Search Algorythms (Algoritmos de Búsqueda)*</l></div>
<div style="text-align:center"> <img src="IMG/DFS & BFS.png" scale=5.0/></div>

---
- `DFS`: Deep-First-Search *(Búsqueda en profundidad)*
- `BFS`: Breadth-First-Search *(Búsqueda en anchura)*

````python
class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)
        
    def dfs_search(self, value):
        """Verifica si un valor está presente en el árbol."""
        if self.name == value:
            return True
        for children in self.children:
            if children.dfs_search(value):
                return True
        return False
        pass
    
    def bfs_search(self, value):
        queue = [self]
        while len(queue) > 0:
            nodo = queue.pop(0)
            if nodo.name == value:
                return True
            for children in nodo.children:
                queue.append(children)
        return False
        pass

    def display(self, depth=0):
        print(' ' * depth * 2 + self.name)
        for child in self.children:
            child.display(depth + 1)

# Creating the tree structure
root = TreeNode("/ (root)")

home = TreeNode("home")
user1 = TreeNode("user1")
user2 = TreeNode("user2")
var = TreeNode("var")
etc = TreeNode("etc")

root.add_child(home)
root.add_child(var)
root.add_child(etc)

# Home subtree
home.add_child(user1)
home.add_child(user2)

# user1 subtree
doc_user1 = TreeNode("documentos")
photo_user1 = TreeNode("fotos")
music_user1 = TreeNode("música")

user1.add_child(doc_user1)
user1.add_child(photo_user1)
user1.add_child(music_user1)

# documentos subtree
work = TreeNode("trabajo")
personal = TreeNode("personal")
studies = TreeNode("estudios")

doc_user1.add_child(work)
doc_user1.add_child(personal)
doc_user1.add_child(studies)

work.add_child(TreeNode("informe1.docx"))
work.add_child(TreeNode("informe2.docx"))
work.add_child(TreeNode("presupuesto.xlsx"))

personal.add_child(TreeNode("carta.txt"))
personal.add_child(TreeNode("recetas.docx"))
personal.add_child(TreeNode("diario.txt"))

studies.add_child(TreeNode("notas.txt"))
studies.add_child(TreeNode("tesis.pdf"))

# fotos subtree
vacation = TreeNode("vacaciones")
events = TreeNode("eventos")

photo_user1.add_child(vacation)
photo_user1.add_child(events)

vacation.add_child(TreeNode("playa.png"))
vacation.add_child(TreeNode("montaña.png"))
vacation.add_child(TreeNode("ciudad.png"))

events.add_child(TreeNode("boda.jpg"))
events.add_child(TreeNode("cumpleaños.jpg"))
events.add_child(TreeNode("navidad.jpg"))

# música subtree
rock = TreeNode("rock")
pop = TreeNode("pop")

music_user1.add_child(rock)
music_user1.add_child(pop)

rock.add_child(TreeNode("album1.mp3"))
rock.add_child(TreeNode("album2.mp3"))

pop.add_child(TreeNode("album1.mp3"))
pop.add_child(TreeNode("album2.mp3"))

# user2 subtree
projects_user2 = TreeNode("proyectos")
downloads_user2 = TreeNode("descargas")
videos_user2 = TreeNode("videos")

user2.add_child(projects_user2)
user2.add_child(downloads_user2)
user2.add_child(videos_user2)

projectA = TreeNode("proyectoA")
projectB = TreeNode("proyectoB")

projects_user2.add_child(projectA)
projects_user2.add_child(projectB)

projectA.add_child(TreeNode("diseño.pdf"))
projectA.add_child(TreeNode("código.py"))

projectB.add_child(TreeNode("presentación.pptx"))
projectB.add_child(TreeNode("datos.csv"))

downloads_user2.add_child(TreeNode("archivo1.zip"))
downloads_user2.add_child(TreeNode("archivo2.zip"))

videos_user2.add_child(TreeNode("película1.mp4"))
videos_user2.add_child(TreeNode("película2.mp4"))

# var subtree
log = TreeNode("log")
tmp = TreeNode("tmp")

var.add_child(log)
var.add_child(tmp)

log.add_child(TreeNode("syslog"))
log.add_child(TreeNode("auth.log"))

tmp.add_child(TreeNode("temp1"))
tmp.add_child(TreeNode("temp2"))

# etc subtree
apache2 = TreeNode("apache2")
ssh = TreeNode("ssh")
network = TreeNode("network")

etc.add_child(apache2)
etc.add_child(ssh)
etc.add_child(network)

apache2.add_child(TreeNode("apache2.conf"))
sites_available = TreeNode("sites-available")
apache2.add_child(sites_available)
sites_available.add_child(TreeNode("default"))
sites_available.add_child(TreeNode("default-ssl"))

ssh.add_child(TreeNode("ssh_config"))
ssh.add_child(TreeNode("sshd_config"))

network.add_child(TreeNode("interfaces"))
network.add_child(TreeNode("resolv.conf"))
````

- `Recorridos, aristas & contenedores`:

Modelar un `Grafo` que tenga: 
- como nodo a un estudiante con la cantidad de dinero que tiene encima (si tiene) 
- como arista a otro(s) estudiante(s) con los cuales se prestarían dinero entre si si fuese necesario. 

Y, conociendo esto, si tiene como partida a un *estudiante* (`nodo`) en concreto que necesita una `x` cantidad de *dinero* prestada y se desea encontrar un *camino* entre sus `aristas`(y la sucesion de `aristas` correspondientes a cada una de ellas de forma sucesiva en caso de que sea necesario) que le permita obtener la cantidad de *dinero* que necesita si la tiene directamente, de lo contrario, debe encontrar este camino en búsqueda de exactamente la cantidad de *dinero* que necesita para prestar a su compañero *si es posible*.
 

Para el grafo creado en la anterior clase practica donde habia una arista entre 2 estudiantes si existia confianza entre ambos como para prestarse dinero, modifiquelo para ponderar la relacion entre los nodos, ahora cada arista tendrá un valor numerico que representa el dinero maximo que podrian llegar a prestarse y cada nodo tendrá un valor numerico que representa la cantidad de dinero actual que lleva arriba, un estudiante A quiere averiguar si haciendo que varios se pidan prestados de manera consecutiva entre si puede lograr conseguir una suma x

````python
class Dinero():
    def __init__(self):
        self.grafo = {}
    
    def __str__(self):
        return str(self.grafo)
    
    def agregar_arista(self, persona=(), relacion=(), monto=0):
        dinero_persona = 0 if isinstance(persona, (tuple,set, list))==False else persona[1]
        persona = persona[0] if isinstance(persona, (tuple, set, list)) else persona
        
        dinero_relacion = 0 if isinstance(relacion, (tuple,set, list))==False else relacion[1]
        relacion = relacion[0] if isinstance(relacion, (tuple,set, list)) else relacion

        if persona not in self.grafo:
            self.grafo[persona] = []
        if relacion not in self.grafo:
            self.grafo[relacion] = []   
        self.grafo[persona].append((relacion, dinero_relacion, monto))
        self.grafo[relacion].append((persona, dinero_persona, monto))

    def prestar_dinero(self, person, monto):
        def dfs(person, restante, recorrido):
            if restante < 0:
                return None 
            if restante == 0:
                return recorrido
            for vecino, dinero, prestable in self.grafo.get(person, 0):
                if vecino not in recorrido:
                    result = dfs(vecino, restante - prestable, recorrido + [vecino])
                    if result is not None:
                        return result
            return None
        
        return dfs(person, monto, [person])

grafo = Dinero()
grafo.agregar_arista(("Luis Ernesto", 200), ("Guillermo", 300), 100)
grafo.agregar_arista(("Guillermo", 300), ("Miguel", 50), 50)
grafo.agregar_arista(("Miguel", 50), ("Alberto", 400), 25)
grafo.agregar_arista(("Alberto", 400), ("Yulia", 400), 300)
grafo.agregar_arista(("Luis Ernesto", 200), ("Yulia", 400), 150)
grafo.agregar_arista(("Yulia", 400), ("Dalia", 150), 150)
grafo.agregar_arista(("Dalia", 150), ("Miguel", 50), 50)
grafo.agregar_arista(("Cristina", 500), ("Alberto", 400), 300)

ruta = grafo.prestar_dinero("Luis Ernesto", 200)
print(grafo)
print(ruta)
````

Modelar un `Grafo` que represente la *temática musical* de una fiesta (géneros musicales) y que contenga:

- como `nodos` las *personas* que irán a la fiesta con sus *intereses musicales* contenidos
- como `aristas` al momento de agregar a una persona se relacionan automaticamente de forma tal que habrá una conexión entre 2 nodos si comparten *al menos un género musical*

Objetivo: Encontrar el grupo de nodos conectados mas grande del grafo


````python
class Nodo:
    def __init__(self,valor,gustos=[]):
        self.valor = valor
        self.gustos = gustos

class Fiesta:
    def __init__(self):
        self.grafo = {}

    def __str__(self):
        return str(self.grafo)
    
    def agregar_nodo(self, nodo):
        if nodo not in self.grafo:
            self.grafo[nodo.valor] = []
        else: 
            raise KeyError("Este nodo ya se encuentra dentro del grafo")        

    def agregar_arista(self, nodo, destino):
        if nodo.valor in self.grafo:
            if destino.valor not in self.grafo[nodo.valor]:
                if len(set(nodo.gustos).intersection(set(destino.gustos))) > 0:
                    self.grafo[nodo.valor].append((destino.valor, destino.gustos))
                    self.grafo[destino.valor].append((nodo.valor, nodo.gustos)) 
        else:
            raise KeyError("La clave insertada no se encuentra dentro del grafo")
    
    def bfs(self, nodo_inicial):
        visitados = []
        cola = deque([nodo_inicial])
        max_grupo = []
        while cola:
            nodo_actual = cola.popleft()
            if nodo_actual not in visitados:
                visitados.append(nodo_actual)
                max_grupo.append(nodo_actual)
                cola.extend(nodo_actual)
                for hijo in self.grafo[nodo_actual]:
                    cola.extend(hijo)
        return max_grupo
    
Tropical = Fiesta()
Luis = Nodo("Luis", ["Repartero", "Electronica", "Rock", "Rap"])
Guillermo = Nodo("Guillermo", ["Electronica", "Rock", "Metall", "Deathcore"])
Miguel = Nodo("Miguel", ["Romantico","Baladas", "Rap", "Pop"])
Dalia = Nodo("Dalia", ["Clasic", "Disco", "Trap"])
Yulia = Nodo("Yulia", ["Repartero", "Trap", "Romantico"])
Cristina = Nodo("Cristina", ["Electronica", "Rock", "K-Pop", "Pop"])
Alberto = Nodo("Alberto", ["Repartero", "Rock","Rap"])

Tropical.agregar_nodo(Luis)
Tropical.agregar_nodo(Guillermo)
Tropical.agregar_nodo(Miguel)
Tropical.agregar_nodo(Dalia)
Tropical.agregar_nodo(Yulia)
Tropical.agregar_nodo(Alberto)
Tropical.agregar_nodo(Cristina)

Tropical.agregar_arista(Luis, Guillermo)
Tropical.agregar_arista(Luis, Miguel)
Tropical.agregar_arista(Luis, Dalia)
Tropical.agregar_arista(Luis, Yulia)
Tropical.agregar_arista(Luis, Alberto)
Tropical.agregar_arista(Luis, Cristina)
Tropical.agregar_arista(Guillermo, Miguel)
Tropical.agregar_arista(Guillermo, Dalia)
Tropical.agregar_arista(Guillermo, Yulia)
Tropical.agregar_arista(Luis, Alberto)
Tropical.agregar_arista(Luis, Cristina)

print(Tropical)
````
